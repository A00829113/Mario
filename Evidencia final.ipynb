```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

Lo que hice en este momento fue importar diferentes liberarias para poder hacer tablas dinamicas, modelos de regresión y diferentes graficas como el histograma,grafica de puntos, graficas de dispersion etc.

```python
df=pd.read_csv('pokemon.csv')
df.head()
```

En los ultimos 3 codigos podemos observar que use la librerias pandas, para importar una base datos la cual se llama pokemon,
en esta podemos observar que al escribir df.head() nos da los primeros 5 renglones y todas sus columnas de esta base de datos,
en df.shape es para ver cuantos renglones y columnas tiene, como podemos ver tiene 13 columnas y 800 renglones. El df.columns nos sirver para ver cuales columnas tenemos.

```python
df.drop(['Sp. Atk','Sp. Def'], axis=1, inplace=True)
print(df.columns)
df.head(2)
```

```python
df.rename(columns={'Name':'Nombre',
 'Type 1':'Tipo 1',
'Type 2':'Tipo 2',
 'HP':'Vida',
 'Attack':'Ataque',
 'Defense':'Defensa',
 'Speed':'Velocidad',
 'Generation':'Generación',
 'Legendary':'Legendario'}, inplace=True )

df.columns
```

Lo que hice en esta parte fue primero, eliminar las columnas que no usare, usando el axis=1 que significa columas, al correr el df.columns se puede observar que se eliminaron esas columnas, lo siguiente que hice fue cambiar los nombres de cada columna a español.

```python
pd.pivot_table(df, index='Generación',
 columns='Tipo 1',
 values='Nombre',
 aggfunc='count')
```

Ahorita hice una tabla dinamica la cual separa por generación cuantos tipos de pokemon hay, lo que hice fue poner como indice la generación, en las columnas puse tipo 1 , de valores cada nombre
y depediendo el nombre ya iba contando de que tipo era.

```python
sns.lmplot(x='Vida',
 y='Total',
 data=df,
 fit_reg=False,
 hue='Legendario')
```

En este paso hice una grafica de dispersion, en la cual separe los pokemones legendario y no legendario poniendo los ejes el de y como Total y el de x como vida, se puede observar que los legendarios su total esta arriba de la media de los no legendarios, pero la vida de los no legendarios es mayor.

```python
sns.swarmplot(x='Generación',
 y='Total',
 data=df)
```

En esta gráfrica fue una de enjambre la cual nos dice que generación es la mas fuerte que las demas. En este caso la tercera generación es la mas fuete.

```python
from matplotlib.pyplot import figure
figure(figsize=(10,5),dpi=100)
plt.subplot(1,3,1)
plt.hist('Ataque', bins=15,data = df, color = 'g')
plt.ylim([0, 140])
plt.xlim([0, 200])
plt.xlabel('Ataque')
plt.ylabel('Frecuencia')
plt.title('Ataque')
plt.subplot(1,3,2)
plt.hist('Defensa', bins=15,data = df,color= 'b')
plt.ylim([0, 175])
plt.xlim([0, 250])
plt.xlabel('Defensa')
plt.ylabel('Frecuencia')
plt.title('Defensa')
plt.subplot(1,3,3)
plt.hist('Velocidad', bins=15,data = df, color= 'y')
plt.ylim([0, 145])
plt.xlim([0, 200])
plt.xlabel('Velocidad')
plt.ylabel('Frecuencia')
plt.title('Velocidad')
plt.show()
```

Ahorita junte tres histogramas en una figura para identificar la frecuencia de cada uno segun su ataque, defensa y velocidad.

Ahorita hice una matriz de covarianza para ver cual tiene una mejor correlación con cada variable

```python
model=sm.OLS.from_formula('Total~Ataque', #BPXSY! =bo +b1+RIDAGEYR
 data=df)
result=model.fit()
result.summary()
```

Al ya tener la regresion, podemos observar que su R2 es .52 es relativamente buena. El modelo es de los minimos cuadrados el coeficiente de x1 es 2.72 y el b0 es 220.13

```python
plt.scatter(y=df['Total'],
 x=df['Ataque'],
 color='red'
 )
plt.xlabel('Ataque')
plt.ylabel('Total')
m,a = np.polyfit(df['Ataque'],df['Total'] , 1)
plt.plot(df['Ataque'], m*df['Ataque'] + a)
```
