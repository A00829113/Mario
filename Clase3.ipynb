```python
df = pd.read_csv('census.csv')
df.head(2)
```

```python
my_pivot = pd.pivot_table(df,
                          index = ['STNAME','CTYNAME'],
                          values = 'BIRTHS2010',
                          aggfunc ='mean', #['mean','sum','median'],
                          columns = 'REGION',
                          margins = True
                         )
my_pivot
```
