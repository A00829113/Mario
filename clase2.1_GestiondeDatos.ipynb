### sintacis de loc es
### [row_initial:row_final, cil_initial:col_final]

nombre de columnas y renglones

```python
ndf = dataFrame.loc[3:10:,'RIDAGEYR']
ndf
```

### sintacis de iloc es
### [row_initial:row_final, cil_initial:col_final]
indices de los renglones y las columnas 

```python
ndf = dataFrame.iloc[3:10:, :10]
ndf
```
